```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/sentiment_results.csv')
print(df.head())
```

```python
plt.figure(figsize=(8, 5))
colors = {'Positive': 'seagreen', 'Negative': 'crimson', 'Neutral': 'silver'}
df['sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=[colors[x] for x in df['sentiment'].value_counts().index])
plt.title('Overall Sentiment Distribution: Gamage Recruiters')
plt.show()
```
